## Model train - 모델 생성

python trainmodel.py "paragraph.txt"

```python
import json
import pprint
import sys
import pandas as pd
from gensim.models import FastText
from konlpy.tag import Kkma

kkma = Kkma()
```

```python
def loadparagraphset(ptxt):
    f = open(ptxt, "r", encoding='utf-8')
    paraset = f.readlines()
    f.close()
    preparaset = []
    for i in paraset:
        for k in i.split("다."):
            if k.replace("\n", "") == '' or k.replace("\n", "") == ' ':
                continue
            preparaset.append(k.replace("\n", "")+"다.")
    return preparaset
```

```python
paraset = loadparagraphset('data.txt')
paraexhyung = []

# 디버깅
for i in paraset:
    print(i)
```

* 보통명사 : 일반적인 사물의 이름을 가리킨다. 예시) 하늘, 나무, 사랑, 희망 등
* 고유명사 : 특정한 사람이나 사물의 이름을 가리킨다. 보통 인명,지명,천체의 이름, 군대의 이름, 신문 이름 등에 많이 쓰인다. 예시) 안창호, 금강산, 신라, 한강, 독도 등

```python
def hyungextrac(text):
    pos = kkma.pos(text)
    morph = []
    for i in pos:
        if i[1] == "NNG" or i[1] == "NNP" or i[1] == "VV": # NNG:보통명사, NNP:고유명사, VV:동사
            morph.append(i[0])
    return morph
```

```python
hyungextrac('노인복지주택에 들어가려는 사람은 임대차계약에 따릅니다.')
```

```python
for i in paraset:
    paraexhyung.append(hyungextrac(i))
```

```python
# 디버깅
for p in paraexhyung:
    print(p)
```

```python
def trainmodel(paragraphset, fs, fw, fc):
    embedding_model = FastText(
        paragraphset, size=fs, window=fw, min_count=fc, workers=4, sg=1)
    mname = str(fs)+"_"+str(fw)+"_"+str(fc)+".model"
    embedding_model.save(mname)
    print(mname+" save done")
    return embedding_model
```

```python
femmodel = trainmodel(paraexhyung, 400, 10, 40)
```

## model test - 테스트 진행

python modeltest.py QA.csv 400_10_40.modelsave

```python
import operator

def QAreader(QAcsv):
    QA = pd.read_csv(QAcsv)
    testdiclst = []
    cate_1 = []
    cate_2 = []
    for i in range(len(QA)):
        if QA['1차 category'][i] not in cate_1:
            cate_1.append(QA['1차 category'][i])
        if QA['2차 category'][i] not in cate_2:
            cate_2.append(QA['2차 category'][i])
        testdiclst.append({"Q": QA['Q'][i], "A": QA['A'][i], "cate1": QA['1차 category']
                           [i], "cate2": QA['2차 category'][i], "paragraph": QA['paragraph'][i]})

    return testdiclst, cate_1, cate_2
```

```python
# 모델 불러오기 
loaded_model = FastText.load('400_10_40.model')
```

```python
# QA 사전, 1차 카테고리, 2차 카테고리
QAdic, cate_1, cate_2 = QAreader('QA.csv')
cc = 0
```

```python
print('QA 1번째 : ', QAdic[0]['Q'], ' - ', QAdic[0]['A'], end='\n\n')
print('cate_1 : ', cate_1)
print('cate_2 : ', cate_2)
```

```python
def calc_similarity(word, model, catelst):
    # word : 질문, model : FastText 모델, catelst : 1차 카테고리
    moon = hyungextrac(word)
    print("word를 형태소 분석한 moon : ", moon)
    
    dislst = {}
    for i in catelst:
        sc = 0
        for t in moon:
            simsc = model.similarity(i, t)  # 하나의 카테고리와 질문의 형태소 별 유사도
            sc += simsc  # 하나의 카테고리와의 유사도 총합
        dislst[i] = sc
    return dislst
```

```python
# 모든 질문,대답 test 및 결과 출력
for i in QAdic:
    temp = calc_similarity(i['Q'], loaded_model, cate_1)
    print("Q : ", i['Q'])
    print("Realcate(원래 1차 카테고리) : ", i['cate1'])
    print("predic : ", max(temp.items(), key=operator.itemgetter(1))[0])  # 가장 유사도가 높은 카테고리
    print("s_matrix", temp)

    if max(temp.items(), key=operator.itemgetter(1))[0] in i['cate1']:
        cc += 1
    print()

print("Accuracy:", cc/len(QAdic))
```
